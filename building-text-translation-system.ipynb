## Tokenizer and Model

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
text = "Ш­Щ‚ Ш§Щ„ЩЃЩ„ШіШ·ЩЉЩ†ЩЉЩЉЩ† ЩЃЩЉ Ш§Щ„ШЇЩЃШ§Ш№ Ш№Щ† ШЈЩ†ЩЃШіЩ‡Щ… Щ‡Щ€ Ш¬ШЁЩ‡Ш© ШЈШіШ§ШіЩЉШ© Щ…Щ† ШµШ±Ш§Ш№Щ‡Щ… Щ…Щ† ШЈШ¬Щ„ Ш§Щ„Ш­Ш±ЩЉШ© Щ€Ш§Щ„ЩѓШ±Ш§Щ…Ш©."

tokenizer.src_lang = "ara_Arab"
inputs = tokenizer(
    text,
    return_tensors='pt',
)
```

```python
forced_bos_token_id = tokenizer.convert_tokens_to_ids('eng_latin')

outputs = model.generate(
    **inputs,
    forced_bos_token_id=forced_bos_token_id, 
    max_new_tokens=256,
    num_beams=5,
    no_repeat_ngram_size=3,
    early_stopping=True
)
```

```python
outputs
```

```python
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print("Translation:", translated_text)
```

```python
def translate(text, src='ara_Arab', tgt='eng_latin'):
    tokenizer.src_lang = src
    inputs = tokenizer(
        text,
        return_tensors='pt'
    )

    forced_bos_token_id = tokenizer.convert_tokens_to_ids(tgt)

    outputs = model.generate(
        **inputs,
        forced_bos_token_id=forced_bos_token_id, 
        max_new_tokens=256,
        num_beams=5,
        no_repeat_ngram_size=3,
        early_stopping=True
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
text = 'Ш§Щ†Ш§ ЩѓШ±ЩЉШіШЄЩЉШ§Щ†Щ€ Ш±Щ€Щ†Ш§Щ„ШЇЩ€ . Щ„Ш§Ш№ШЁ ЩѓШ±Ш© Щ‚ШЇЩ… ЩЃЩ‰ Щ†Ш§ШЇЩ‰ Ш±ЩЉШ§Щ„ Щ…ШЇШ±ЩЉШЇ. Ш§Щ†Ш§ ШЁШ±ШЄШєШ§Щ„Щ‰ Ш§Щ„Ш¬Щ†ШіЩЉШ©.'

print(f"Translation: {translate(text)}")
```

## Try Conversational Pipeline

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

# https://huggingface.co/Qwen/Qwen3-0.6B

pipe = pipeline("text-generation", model="Qwen/Qwen3-0.6B")
```

```python
question = 'Who is Cristiano Ronaldo ?'
```

```python
messages = [
    {"role": "user", "content": question},
]

response = pipe(messages)
response
```

```python
response[0]['generated_text']
```

```python
response[0]['generated_text'][1]['content']
```

```python
def request_and_response(question: str) -> str:
    messages = [{
        'role': 'user',
        'content': question
    }]

    return pipe(messages)[0]['generated_text'][1]['content']
```

```python
request_and_response("Compute: 100 + 312 = ??")
```

```python
request_and_response("What is the Price of Golds today?")
```

```python
request_and_response("What is the Machine Learning ?")
```

```python
prompt = "Act like an AI Engineer and Give me the response in a bullet points. What is the Machine Learning ?"
request_and_response(prompt)
```

## Asking Follow-up Questions¶
- Finally, let's ask the model follow-up questions to see how it will respond to them.
- We will keep your previous prompt as well as the LLM's response so that the LLM will converse with you as if it remembers your earlier conversation.
- We will ask what else you recommend:

```python
conversation_history = []

def add_message_and_respond(history, new_message):
    '''
        history: history of conversation to be like a context for the LLM.
        new_message: user request
    '''
    # Add the new message to the conversation history
    history.append({
        'role':'user',
        'content': new_message
    })

    # Combine all previous messages for context
    full_conversation = ' '.join([msg['content'] for msg in history])

    # Get a response considering the full conversation
    response = pipe(full_conversation,  max_new_tokens=256)
    bot_response_text = response[0]['generated_text']

    if bot_response_text.startswith(full_conversation):
        bot_response_text = bot_response_text[len(full_conversation):].strip()

    # Add Bot's response to the history
    history.append({
        'role': 'assistant',
        'content': bot_response_text
    })

    return bot_response_text
```

```python
# Start the conversation
initial_message = "What are the best ways to learn AI?"

response = add_message_and_respond(conversation_history, initial_message)

print("Bot:", response)
```

```python
# Continue the conversation
follow_up_message = "What else do you recommend?"

response = add_message_and_respond(conversation_history, follow_up_message)

print("Bot:", response)
```

```python
# Continue the conversation
follow_up_message = "Do you recommend certain book about machine learning?"

response = add_message_and_respond(conversation_history, follow_up_message)

print("Bot:", response)
```

## Try Tokenizer and Model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-0.6B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen3-0.6B")
```

```python
question = 'Who are you?'
messages = [
    {"role": "user", "content": question},
]

messages
```

```python
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)

print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
def ask_qwen(question):
    messages = [{
        'role': 'user',
        'content': question
    }]

    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        tokenize=True,
        return_dict=True,
        return_tensors="pt"
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=48
    ) 

    return tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)
```

```python
ask_qwen("What is AWS ?")
```
